```python
import pandas as pd
import scipy.stats as stats

# 1. Loading the dataset
df = pd.read_csv(r"C:\Users\nithish\Downloads\Task-4(Apex)\Sample - Superstore.csv", encoding="latin1")
print("Dataset loaded successfully!")
print(f"Total records found: {len(df)}\n")

# 2. Separating  groups using the actual columns: 'Segment' and 'Sales'
# comparing the Consumer group sales vs Corporate group sales
consumer_sales = df[df['Segment'] == 'Consumer']['Sales']
corporate_sales = df[df['Segment'] == 'Corporate']['Sales']

print(f"Number of Consumer orders: {len(consumer_sales)}")
print(f"Number of Corporate orders: {len(corporate_sales)}")

# 3.  Welch's T-Test
t_stat, p_value = stats.ttest_ind(consumer_sales, corporate_sales, equal_var=False)

# 4. Print the final results
print("\n=== RAW STATISTICAL TEST RESULTS ===")
print(f"Calculated T-statistic: {t_stat:.4f}")
print(f"Calculated P-value:     {p_value:.6f}")
print("====================================")

if p_value < 0.05:
    print("RESULT: Statistically Significant! Reject Null Hypothesis. Order values differ between segments.")
else:
    print("Conclusion: Not Significant. Fail to reject Null Hypothesis. Observed differences are purely random.")
```
